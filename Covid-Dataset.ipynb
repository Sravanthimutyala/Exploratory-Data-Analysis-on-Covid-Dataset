```python
# importing the Necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Set seaborn style
sns.set(style='whitegrid')
```

```python
# Loading the dataset

df = pd.read_csv("Downloads/country_wise_latest.xls")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Shape of the Row:\n",df.shape[0])
print("Shape of the Column:\n", df.shape[1])
```

```python
df.isnull().sum()
```

```python
focus_cols = [
    'Country/Region', 'WHO Region',
    'Confirmed', 'Deaths', 'Recovered', 'Active',
    'New cases', 'New deaths', 'New recovered',
    'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered'
]

df = df[focus_cols]
df.head()
```

```python
df.columns
```

```python
top_new_deaths = df.sort_values(by='New deaths', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_new_deaths, x='New deaths', hue='Country/Region',legend = True,palette = "Greens")
plt.title('Top 10 Countries by New Deaths')
plt.xlabel('New Deaths')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
top_recovery_rate = df.sort_values(by='Recovered / 100 Cases', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_recovery_rate, x='Recovered / 100 Cases', hue='Country/Region', palette='Reds')
plt.title('Top 10 Countries by Recovery Rate')
plt.xlabel('Recovery Rate (%)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df, 
    x='Recovered', 
    y='Deaths', 
    hue='WHO Region', 
    size='Confirmed', 
    alpha=0.7
)
plt.title('Scatter Plot: Deaths vs Recoveries')
plt.xlabel('Recovered Cases')
plt.ylabel('Deaths')

plt.tight_layout()
plt.show()
```

```python


plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Key Metrics')
plt.tight_layout()
plt.show()
```

```python
# Daily Cases

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='', y='New cases', hue='Country/Region')
plt.title('Daily New COVID-19 Cases by Country')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
